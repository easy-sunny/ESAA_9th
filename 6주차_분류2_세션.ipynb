3. 변수를 변환합니다.

```python
# type에는 white와 red 두 종류가 있습니다.
# 각각 0,1로 변환합니다.
type2=[]
for i in train["type"]:
    if i == "white":
        type2.append(0)
    else:
        type2.append(1)
```

```python
type2=[]
for i in test["type"]:
    if i == "white":
        type2.append(0)
    else:
        type2.append(1)
```

```python
test.head()
```

```python
train.drop("type",inplace=True, axis=1)
test.drop("type",inplace=True, axis=1)
```

```python
train.rename(columns={"type2":"type"},inplace=True)
test.rename(columns={"type2":"type"},inplace=True)
```

```python
train.head()
```

```python
test.head()
```

```python
# 모델에 입력하기 전 데이터를 정형합니다.
# train_x : train의 'index', 'quality' 피처를 제거하고 저장합니다.
# train_y : train의 'quality'를 저장합니다.
# test_x : test의 'index'를 제거하고 저장합니다.

train_x = train.drop(["index","quality"],axis=1)
train_y = train["quality"]
test_x = test.drop("index",axis=1)
```

```python
train_x.shape, train_y.shape, test_x.shape
```

```python
train_x.head()
```

```python
train_y.head()
```

```python
from lightgbm import LGBMClassifier
lgbm_wrapper = LGBMClassifier(n_estimators=400)

lgbm_wrapper.fit(train_x, train_y)
```

5. 생성한 모델로 예측 데이터를 생성합니다.

```python
preds = lgbm_wrapper.predict(test_x)
```

6. 제출파일을 생성합니다.

```python
smpl_sub.head()
```

```python
#서브미션 파일의 quality 변수에 예측값을 대입합니다.

smpl_sub["quality"]= preds
```

```python
# csv 파일로 저장합니다.
smpl_sub.to_csv('submission.csv', index=False)
```

# 3. 보팅

```python
# 데이터 다운로드 링크로 데이터를 코랩에 불러옵니다.

!wget 'https://bit.ly/3i4n1QB'

import zipfile
with zipfile.ZipFile('3i4n1QB', 'r') as existing_zip:
    existing_zip.extractall('data')
```

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier

from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier

train1 = pd.read_csv('data/train.csv')
test1 = pd.read_csv('data/test.csv')

train1.drop('index',axis = 1 ,inplace =True)
test1.drop('index',axis = 1 ,inplace =True)
```

```python
# 데이터 확인
train1.info()
```

```python
# 원핫 인코딩 (pd.get_dummies())

train1=pd.get_dummies(train1,columns=["type"])
test1=pd.get_dummies(test1,columns=["type"])
```

```python
# 모델 정의 (LightGBM, XGBoost, Random Forest)


from lightgbm import LGBMClassifier
lgbm_wrapper = LGBMClassifier(n_estimators=400)

from xgboost import XGBClassifier
xgb_wrapper = XGBClassifier(n_estimators=400, learning_rate=0.1, max_depth=3)

from sklearn.ensemble import RandomForestClassifier
rt_clf = RandomForestClassifier()

# VotingClassifier 정의
from sklearn.ensemble import VotingClassifier
vo_clf = VotingClassifier(estimators =[("lgbm",lgbm_wrapper),("xgb",xgb_wrapper),("rfc",rt_clf)], voting="soft")
```

```python
# 모델 학습

# X 는 train에서 quality 를 제외한 모든 변수
# y 는 train의 qulity 변수

X = train1.drop('quality',axis= 1)
y = train1['quality']

# fit 메소드를 이용해 모델 학습
vo_clf.fit(X,y)
```

```python
# predict 메소드와 원핫 인코딩한 테스트 데이터를 이용해 품질 예측
vo_clf.predict(test1)
```
