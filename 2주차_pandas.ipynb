# 판다스

판다스는 파이썬에서 데이터처리를 위해 존재하는 가장 인기 있는 라이브러리<br>
일반적으로 데이터 세트는 2차원 데이터 (행_row X 열_column) -> 판다스는 2차원 데이터를 효율적으로 가공/처리할 수 있음

판다스의 핵심 객체<br> 
DataFrame: 여러 개의 행과 열로 이루어지는 2차원 데이터를 담는 데이터 구조체<br>
Series: 칼럼이 하나뿐인 데이터 구조체 (반면, DataFrame은 칼럼이 여러 개인 데이터 구조체)<br>
Index: 개별 데이터를 고유하게 식별하는 Key값<br>
=> Series와 DataFrame은 모두 Index를 key 값으로 가지고 있음<br>


## 판다스 시작 - 파일을 DataFrame으로 로딩, 기본 API

판다스는 다양한 포맷으로 된 파일을 DataFrame으로 로딩할 수 있는 편리한 API를 제공

: CSV(칼럼을 ','로 구분한 파일 포맷)파일 포맷변환을 위한 API<br>
: 사실 CSV뿐만 아니라 어떤 필드 구분 문자 기반의 파일 포맷도 DataFrame으로 변환이 가능<br>

read_fwf()<br>
: Fixed Width, 즉 고정 길이 기반의 칼럼 포맷을 DataFrame으로 로딩하기 위한 API

=> 일반적으로 DataFrame으로 변환 시, 이 칼럼명을 지정해 줍니다. (지정하지 않으면 자동으로 칼럼명 할당)

판다스 DataFrame 객체<br>
생성인자 data는 리스트, 딕셔너리, 넘파이 ndarray를 입력 받음<br>

DataFrame은 기본적으로 행과 열을 가지는 2차원 데이터 = 2차원 이하의 데이터들만 DataFrame으로 변환 가능

```python
import numpy as np
```

```python
col_name1 = ["col1"]

list1 = [1,2,3]
array1 = np.array(list1)
print("array1 shape:", array1.shape)
```

```python
#리스트를 이용해 DataFrame 생성
df_list1 = pd.DataFrame(list1, columns=col_name1)
print("1차원 리스트로 만든 DataFrame:\n", df_list1)
```

```python
#넘파이를 이용해 DataFrame 생성
df_array1 = pd.DataFrame(array1, columns=col_name1)
print("1차원 ndarray로 만든 DataFrame:\n", df_array1)
```

```python
#2행X3열 형태의 리스와 ndarray 생성한 뒤 이를 Data Frame으로 변환 = 3개의 칼럼명이 필요함

col_name2 = ["col1","col2","col3"]

list2 = [[1,2,3],
        [11,12,13]]

array2 = np.array(list2)

print("array2 shape:", array2.shape)
print("---------------------------")

df_list2 = pd.DataFrame(list2, columns=col_name2)
print("2차원 리스트로 만든 DataFrame:\n", df_list2)
print("---------------------------")

df_array2 = pd.DataFrame(array2, columns=col_name2)
print("2차원 ndarray로 만든 DataFrame:\n", df_array2)
```

## DataFrame 삭제

#### drop( ) 메서드

DataFrame.drop(**labels=None**, **axis=0**, index=None, level=None, **inplace=False**,errors="raise")

가장 중요한 파라미터는 labels, axis, inplace

- axis <br>
: 특정 칼럼 또는 특정 행을 드롭.<br>
2차원 ndarray의 구성 시 axis0과 axis1이 있음을 적용 = 판다스의 DataFrame은 2차원 데이터만 다루므로 axis0, axis1<br>
=> axis=1을 입력 ; labels에 원하는 칼럼 명 입력하면 지정된 칼럼을 드롭<br>
=> axis=0을 입력; 특정 로우를 드롭하겠다는 뜻. **DataFrame에서 특정 로우를 가리키는 것은 인덱스** labels에 오는 값을 인덱스로 간주

- inplace<br>
inplace는 디폴트 값이 Fasle이므로 inplace 파라미터를 기재하지않으면 자동으로 False<br>
inplace = False ; 자기 자신의 DataFrame의 데이터는 삭제하지 않으며, 삭제된 결과 DataFrame을 반환<br>
inplace = True ; 자신의 DataFrame의 데이터를 삭제<br>
; 이때 유의할 점은 반환 값이 none이 됨.<br>
따라서 inplace=True로 설정한 채로 반환 값을 다시 자신의 DataFrame 객체로 할당하면 안됨.<br> 
titanic_df= titanic_df.drop(["Age_0","Age_by_10","Family_No"],axis=1,inplace=True)는 titanic_df 객체 변수를 아예 none으로 만들어 버림

또한 여러 개의 칼럼을 삭제하고 싶으면 리스트 형태로 삭제하고자 하는 칼럼 명을 입력해 labels 파라미터로 입력하면 됨. 

```python
titanic_drop_df = titanic_df.drop("Age_0",axis=1)
titanic_drop_df.head(3)
```

```python
titanic_df.head(3)
```

앞의 코드에서 inplace = False가 자동으로 지정되었기 때문에 원본 Titanic DataFrame에는 삭제됐다고 생각했던 "Age_0" 칼럼이 여전히 존재 

```python
drop_result= titanic_df.drop(["Age_0","Age_by_10","Family_No"],axis=1,inplace=True)
print("inplace=True로 drop 후 반환된 값:", drop_result)
```

#### reset_index( ) 메서드 

: 새롭게 인덱스를 연속 숫자 형으로 할당하며 기존 인덱스는 "index"라는 새로운 칼럼 명으로 추가됨<br>
: reset_index()는 인덱스가 연속된 int 숫자형 데이터가 아닐 경우에 다시 이를 연속 int 숫자형 데이터로 만들 때 주로 사용<br>
ex) "Pclass" 칼럼 Series의 value_counts()를 수행하면 "Pclass" 고유 값이 식별자 인덱스 역할을 함<br>
이보다는 연속 숫자형 인덱스가 고유 식별자로 더 적합해보 임 -> "reset_index()" 이용<br>
: Series에 reset_index()를 적용하면, 기존 인덱스가 칼럼으로 추가돼 칼럼이 2개 => DataFrame이 반환<br>
: reset_index()의 parameter 중 drop=True로 설정하면 기존 인덱스는 새로운 칼럼으로 추가되지 않고 삭제(drop)됨<br>
=새로운 칼럼이 추가되지 않으므로 그대로 series로 유지됨 (dataframe으로 반환 X)

```python
titanic_reset_df = titanic_df.reset_index(inplace=False)
titanic_reset_df.head(3)
```

```python
print("###Before reset_index###")

value_counts = titanic_df["Pclass"].value_counts()
print(value_counts)
print("value_counts 객체 변수 타입:", type(value_counts))

new_value_counts = value_counts.reset_index(inplace=False)
    
print("###After reset_index###")
print(new_value_counts)
print("new_value_counts 객체 변수 타입:", type(new_value_counts))
```

## 데이터 셀렉션 및 필터링

넘파이의 경우 "[ ]" 연산자 내 단일 값 추출, 슬라이싱, 팬시 인덱싱, 불린 인덱싱을 통해 데이터 추출<br>
판다스의 경우 ix[ ], iloc[ ], loc[ ] 연산자를 통해 동일한 작업 수행


**위치 기반 인덱싱과 명칭 기반 인덱싱의 구분**<br>
-위치 기반 인덱싱<br>
: 0을 출발점으로 하는 가로축, 세로축 좌표 기반의 행과 열 위치를 기반으로 데이터를 지정.<br>
: 행, 열 값으로 정수가 입력됨.<br> 
-명칭 기반 인덱싱<br>
: 칼럼의 명칭을 기반으로 위치를 지정하는 방식<br>
: '칼럼 명' 같이 명칭으로 열 위치를 지정하는 방식


### DataFrame의 [ ] 연산자<br>
: 칼럼만 지정할 수 있는 칼럼지정 연산자<br>
넘파이에서 [ ] 연산자는 행의 위치, 열의 위치, 슬라이싱 범위 등을 지정해 데이터를 가져올 수 있음<br>
하지만 DataFrame 바로 뒤에 있는 [ ] 안에 들어갈 수 있는 것은 칼럼 명 문자 또는 칼럼 명의 리스트 객체<br>
: 인덱스로 변환 가능한 표현식<br>

### 판다스 DataFrmae의 lambda 식 

lambda 식은 if else를 지원<br>
주의할 점_1: if 절의 경우 if 식보다 반환 값을 먼저 기술해야 함.<br>
왜냐하면 lambda 식 ":" 기호의 오른편에 반환 값이 있어야 하기 때문. else의 경우는 else 식이 먼저 나오고 반환 값이 나중에 오면 됨<br>
주의할 점_2: if, else만 지원하고, if-else if-else는 지원하지 않음.<br>
else 절을 ()로 내포해 () 내에서 다시 if else를 적용해 사용

```python
titanic_df["Name_len"] = titanic_df["Name"].apply(lambda x: len(x))
titanic_df[["Name","Name_len"]].head(3)
```

```python
#나이가 15세 미만이면 "Child", 그렇지 않으면 "Adult"로 구분하는 새로운 칼럼 "Child_Adult"

titanic_df["Child_Adult"] = titanic_df["Age"].apply(lambda x : "Child" if x <=15 else "Adult")
titanic_df[["Age","Child_Adult"]].head(8)
```

```python
#나이가 15세 미만이면 "Child", 15세에서 60세는 "Adult", 61세 이상은 "Elderly" 구분하는 새로운 칼럼 "Age_cat"

titanic_df["Age_cat"]=titanic_df["Age"].apply(lambda x: "Child" if x<=15
                                              else ("Adult" if x<=60 else "Elderly"))

titanic_df["Age_cat"].value_counts()
```

else if가 많이 나와야 하는 경우나 switch case문의 경우는 이렇게 else를 계속 내포해서 쓰기 부담<br>
-> 별도의 함수를 만드는 게 더 나음 

```python
#나이에 따라 세분화된 분류를 수행하는 함수 생성
def get_category(age):
    cat=""
    if age <=5: cat="Baby"
    elif age <=12: cat="Child"
    elif age <=18: cat="Teenager"
    elif age <=25: cat="Student"
    elif age <=35: cat="Young Adult"
    elif age <=60: cat="Adult"
    else: cat="Elderly"
        
    return cat
```

```python
#lambda식에 위에서 생성한 get_category()함수를 반환값으로 지정
#get_category(x)는 입력값으로 "Age" 칼럼 값을 받아서 해당하는 cat을 반환

titanic_df["Age_cat"]=titanic_df["Age"].apply(lambda x : get_category(x))
titanic_df[["Age","Age_cat"]].head()
```
